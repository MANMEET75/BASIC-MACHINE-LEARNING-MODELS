```python
data['sex'].value_counts()
```

```python


sex_mapping={"male":0,"female":1}
data['sex']=data['sex'].map(sex_mapping)
```

```python
data['smoker'].value_counts()
```

```python
smoker_mapping={"yes":1,"no":0}
data['smoker']=data['smoker'].map(smoker_mapping)
```

```python
data['region'].value_counts()
```

```python
region_mapping={"southeast":0,"southwest":1,"northwest":2,"northeast":3}
data['region']=data['region'].map(region_mapping)
```

```python
data
```

```python
X=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from xgboost import XGBRegressor
```

```python

# from sklearn.ensemble import GradientBoostingRegressor

# regressor=LinearRegression()
# regressor=KNeighborsRegressor() # giving 83% accuracy over here
regressor=XGBRegressor()


regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict((X_test))
np.set_printoptions(precision=2)
print(np.concatenate((y_test.reshape(len(y_test),1),y_pred.reshape(len(y_pred),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
